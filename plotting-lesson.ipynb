## Plotting Lesson

```python
import numpy
import os
```

```python
distance_file = os.path.join('data','distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file,delimiter=',',dtype='unicode')
print(distances)
```

```python
headers = distances[0]
data = distances[1:]
data = data.astype(numpy.float)
```

```python
print(headers)
print(data)
```

```python
# import matplotlib.pyplot 
# using a function from this library it's libraryname.functionname, so for this one it is matplotlib.pyplot.function
# can give libraries nicknames
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(data[:,1]) # just has y points, x is the count of the rows
# plt.plot(data[:,0],data[:,1]) will give the same plot since the first column is the frame number
```

```python
sample = headers[1]
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
fig_1 = plt.plot(data[:,0],data[:,1], label=sample)
plt.legend()

plt.savefig(F'{sample}.png')
# can also set resolution of the file using dpi=resolution
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,0],data[:,1], label = headers[1])
plt.plot(data[:,0], data[:,2], label = headers[2])
plt.legend()
```

```python
for col in range(1, len(headers)):
    plt.figure()
    fig = plt.plot(data[:,0], data[:,col], label = headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstroms)')
    plt.legend()
    plt.savefig(F'{headers[col]}.png', dpi=300)
```

```python
print(data[0:10:2,:]) # range of rows is 0 to 10, skip every other one
```

```python
plt.figure()
plt.plot(data[0::100,0],data[0::100,1],'--ok') # start at zero, go to the end, plot every 100 points
# '--' dashed lines, 'o' points, '-o' connected pionts, '--o' dashed connected points, check help for more
# the intervals must be the same step size for both x and y
# matplotlib has a ton of plot types on the websitet
```
